```python
#Importando biblioteca pandas:
import pandas as pd
```

```python
#Importando a base de dados:
base_clientes = pd.read_csv('base_clientes.csv', sep = ";")
```

```python
#Conferindo o formato dos dados:
base_clientes.head()
```

```python
#Definindo uma função para marcar as flags de cada produto:
def marcar_flags(flag_produto):
    produto = flag_produto['PRODUTO'].lower()

    #Caso o produto tenha a palavra "star" marque a flag_tv
    if 'star' in produto:
        flag_produto['FLAG_TV'] = 1
    else:
        flag_produto['FLAG_TV'] = 0

    #Caso o produto tenha a palavra "banda larga" marque a flag_banda_larga
    if 'banda larga' in produto:
        flag_produto['FLAG_BANDA_LARGA'] = 1
    else:
        flag_produto['FLAG_BANDA_LARGA'] = 0

    #Caso o produto tenha a palavra "fone" marque a flag_fone
    if 'fone' in produto:
        flag_produto['FLAG_FONE'] = 1
    else:
        flag_produto['FLAG_FONE'] = 0

    return flag_produto

#Aplicando a função em cada linha
base_clientes = base_clientes.apply(marcar_flags, axis=1)

#Salva o arquivo csv novamente com as flags aplicadas
base_clientes.to_csv('base_clientes_tratada.csv', sep=';', index=False)
```
